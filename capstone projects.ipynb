```python
import random
def coin():
    h=0
    t=0
    s=int(input('How many times you have to flip the coin'))
    
    while s>0:
        s=s-1
        a=random.randint(0,1)
        b={0:'head',1:'tail'}
        c=b[a]
        
        if c=='head':
            h+=1
            continue
        else:
            t+=1
            continue
    print(f'no. of heads = {h}')  
    print(f'no. of tails = {t}')  
           
```

```python
coin()
```

```python
def collatz():
    n=int(input('enter a number to check collatz conjecture'))
    step=0
    b=[n]
    
    while n>1:
        
        if n%2==0:
            n=n/2
            step+=1
            b.append(int(n))
            
            continue
         
        else:
            n=n*3+1
            step+=1
            b.append(int(n))
            continue
    print(f'total steps to reach one = {step}')
    print(f'sequence to reach one is {b}')
```

```python
collatz()
```
